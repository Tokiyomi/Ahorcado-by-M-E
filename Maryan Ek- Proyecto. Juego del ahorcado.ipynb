## Proyecto: Juego del ahorcado

```python
print('Bienvenido al juego del ahorcado :)\nPara ello es necesario que conozcas lo siguiente: \n')
print('* Una palabra del idioma español será seleccionada de forma aleatoria.')
print('* Las palabras están en minúscula.\n* El juego termina cuando ganas o pierdes todas tus vidas (tienes 5).')

import time
time.sleep(3)

nombre = input('\n¿Cómo te llamas?: ')

horca = (' -----\n |   |\n     |\n     |\n     |\n     |\n------\n')
ahor1 = (' -----\n |   |\n o   |\n |   |\n     |\n     |\n------\n')
ahor2 = (' -----\n |   |\n\o   |\n |   |\n     |\n     |\n------\n')
ahor3 = (' -----\n |   |\n\o/  |\n |   |\n     |\n     |\n------\n')
ahor4 = (' -----\n |   |\n\o/  |\n |   |\n  \  |\n     |\n------\n')
ahorcado = (' -----\n |   |\n\o/  |\n |   |\n/ \  |\n     |\n------\n')

#Abro el archivo que contiene las palabras y lo convierto en una lista
lista_palabras = open('palabras_uso_comun_2.txt','r')
lista_palabras = list(lista_palabras)

# Ahora se eliminarán todos los saltos de línea y las palabras serán minúsculas
for i in range(len(lista_palabras)):
    lista_palabras[i] = lista_palabras[i] [0:-1]
     
for i in range(len(lista_palabras)):
    lista_palabras[i] = lista_palabras[i][:].lower()

# Importamos random y guardamos en una variable la palabra aleatoria
import random
secreto = random.choice(lista_palabras)
secreto = list(secreto)

# Para cada letra en la palabra secreta habrá un asterisco
aster = ('*' for i in secreto)
aster = list(aster)

# Empieza el juego
print('\n¿Estás listo?\nVidas = 5\n\nEmpecemos, mucha suerte :)\n')
time.sleep(3)
print(horca)
print("".join(aster), '\n')

abecedario = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

vidas = 5
fallas = 0

while vidas > 0:
    
    if fallas == fallas + 1:
        vidas = vidas - 1
    
    intento = str(input('\nDime la letra: '))
   
    if (len(intento) > 1) or (intento == ''):
        print('Lo siento, tu respuesta es inválida.')
        print(aster)
        fallas = fallas + 1
    elif intento not in abecedario:
        print('Lo siento, ese caracter no es válido, debes ingresar una letra del abecedario.')
        print(aster)
        fallas = fallas + 1
    elif intento not in secreto:
        print('Lo siento, esa letra no corresponde a la palabra :(')
        print(aster)
        fallas = fallas + 1
    elif intento in secreto:
        print('Muy bien hecho, la letra corresponde a la palabra :)')
        
        for indice in range(len(aster)):
            if intento == secreto[indice]:
                aster[indice] = intento        
        print(aster)
    
    if fallas == 0:
        print(horca)
    if fallas == 1:
        print('Vidas = 4')
        print(ahor1)
    if fallas == 2:
        print('Vidas = 3')
        print(ahor2)
    if fallas == 3:
        print('Vidas = 2')
        print(ahor3)
    if fallas == 4:
        print('Vidas = 1')
        print(ahor4)
    if fallas == 5:
        vidas = 0
        time.sleep(2)
        print('Vidas = 0\n')
        print(nombre,' has perdido el juego :(\n¡Ánimo, la próxima vez lo harás mejor!')
        print('\nLa palabra era:',''.join(secreto))
        print(ahorcado)
    if aster == secreto:
        print('¡Felicidades!',nombre,' has jugado excelente y ganaste :)')
        break
```
